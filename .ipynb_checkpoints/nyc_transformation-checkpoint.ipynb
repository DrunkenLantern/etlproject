```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import pymongo

data_path = 'assets/ny_restaurant_data.csv'
```

```python
Data_clean = pd.read_csv(data_path)
Data_clean.columns = Data_clean.columns.str.replace(' ','_')

Data_clean
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
db = client.Restaurants_db
collection = db.items
```

```python
reduced_df = Data_clean[["Company_Name", "Sub_Subindustry", "Address"]]
```

```python
renamed_df=reduced_df.rename(columns={"Company_Name": "Name",  "Sub_Subindustry": "Type"})
```
